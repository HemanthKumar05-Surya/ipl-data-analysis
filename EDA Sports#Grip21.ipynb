##### Finding out the most successful teams,it means who won most of the matches

```python
successful_teams=matches['winner'].value_counts()
successful_teams.head(n=10)
```

```python
#Visualizing using a count plot
sns.countplot('winner', data=matches,palette="Paired",linewidth=2,edgecolor=sns.color_palette('dark',4))
plt.xlabel('Team Names')
plt.ylabel('winner Count')
plt.title('Team Names vs Winner Count',fontsize=30)
sns.set_theme(style="darkgrid")
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(13,9)
plt.show()
```

#### Now visualizing winning teams by most of the runs

```python
import seaborn 
```

```python
ax=seaborn.stripplot(x='winner',y='win_by_runs',data=matches,jitter=0.1,palette="Set2",linewidth=2,edgecolor='green')
plt.xlabel('Winning Teams')
plt.ylabel('win by runs')
plt.title('Winning teams by most runs',fontsize=30)
seaborn.set(style = 'darkgrid')   
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(13,6)
plt.show()
```

#### Players contributing for the winning of the team,checking by the player_of_match trophy maximum times

```python
matches['player_of_match'].value_counts().head(n=10)
```

```python
# Creating color parameters 
colors = ( "orange", "cyan", "brown","grey", "indigo", "beige","red","yellow")
# Wedge properties 
wp = { 'linewidth' :1.5, 'edgecolor' : "black" } 

# Creating autocpt arguments 
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%\n({:d} g)".format(pct, absolute) 

labels='CH Gayle','AB de Villiers','MS Dhoni','YK Pathan','SR Watson','SK Raina','G Gambhir','MEK Hussey'
sizes = [21, 20, 17, 16,15,14,13,12]
explode = (0,0.2,0,0,0.2,0,0.3,0.4) 
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',colors=colors,wedgeprops=wp,shadow=True,
        textprops = dict(color ="magenta"),startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

### The lollipop chart is a composite chart with bars and circles. It is a variant of the bar chart with a circle at the end, to highlight the data value. Like a bar chart, a lollipop chart is used to compare categorical data.

```python
#LolliPop Plot-visualizing win by runs
plt.stem(matches.win_by_runs, markerfmt=' ')
(markers, stemlines, baseline) = plt.stem(matches.win_by_runs)
plt.setp(markers,marker='D',linestyle="-",color="olive",linewidth=0.5,markersize=5,markeredgecolor="red",markeredgewidth=1)
seaborn.set(style ='darkgrid')   
fig=plt.gcf()
fig.set_size_inches(18,12)
plt.xticks(rotation=90)
plt.show()
```

### Toss effect on winning the matches

```python
winning_toss_effect=matches.groupby(['toss_winner','toss_decision'])['winner'].agg(['count']).sort_values('count',ascending=False)
winning_toss_effect.head(n=10)
```

### From the above table we can see that KKR won the toss 62 times and they decided to field and RR won the toss for 32 times and decided to bat

#### Checking Venue Effect On Winning

```python
winning_venue=matches.groupby(['venue','winner'])['winner'].agg(['count']).sort_values('count',ascending=False)
winning_venue
```

#### From the above table we can see that KKR won 45 matches at eden gardens and MI won 42 matches at wankhede stadium

#### Checking winning effect by date,runs,wickets and player of the match

```python
winning_final=matches.groupby(['date','win_by_runs','win_by_wickets','player_of_match'])['winner'].agg(['count']).sort_values('count',ascending=False)
winning_final
```

#### Finding out the teams performance by the wickets

```python
matches[matches['win_by_wickets']>0].groupby(['winner'])['win_by_wickets'].apply(np.median).sort_values(ascending=False)
```

```python
fig, ax = plt.subplots()
ax.set_title("Winning by Wickets - Team Performance")
sns.boxplot(y = 'winner', x = 'win_by_wickets', data=matches[matches['win_by_wickets']>0], orient = 'h'); #palette="Blues");
fig=plt.gcf()
fig.set_size_inches(10,6)
plt.show()
```

### Batsman with maximum runs

```python
batsman_maxruns=deliveries.groupby(['batsman'])['batsman_runs'].agg(['count']).sort_values('count',ascending=False)
batsman_maxruns.head(n=10)
```

#### From the above table,we can see that virat kohli has the maximum runs and least is chris gayle
